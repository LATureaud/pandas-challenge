## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
purchase_data.head()

#bins = [0, 9, 14, 19, 24, 29, 34, 39, 500]
#array([ 0,  5, 10, 15, 20, 25, 30, 35, 40, 45, 50], dtype=int64)

groups2 = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]
max_age = purchase_data['Age'].max()
bin2 = [0, 9, 14, 19, 24, 29, 34, 39, 500]
bin2

AGES = pd.cut(purchase_data["Age"], bin2, labels=groups2)
purchase_data["Age Group"]= AGES
purchase_data

age_count= purchase_data.groupby(["Age Group"]).count()

sum_groups = purchase_data.groupby(["Age Group"]).sum()

count_purchase = age_count["SN"]
total_purchase= sum_groups["Price"]
avg_purchase = total_purchase/age_count["Price"]

per_person = total_purchase/count_purchase


age_count["Total Expenditure"]= total_purchase.map("${:,.2f}".format)
age_count["Average Purchase Price"]= avg_purchase.map("${:,.2f}".format)
age_count["Average Expenditure Per Player"]= per_person.map("${:,.2f}".format)
age_count=age_count[["SN", "Total Expenditure", "Average Purchase Price", "Average Expenditure Per Player"]]
age_count= age_count.rename(columns={"SN":"Total Purchases"})
age_count
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
total_purchase = purchase_data.groupby(["SN"]).sum()
big = total_purchase ["Price"].sort_values(ascending=False).head(10)
play_name = big.index
top_spenders = purchase_data.loc[purchase_data["SN"].isin(play_name)]

big_count = top_spenders.groupby(["SN"]).count()
big_sum = top_spenders.groupby(["SN"]).sum()

spend_count = big_count['Age']
total_spend = big_sum['Price']
avg_spend = total_spend/spend_count

top_analysis = pd.DataFrame({"Number of Purchases": spend_count,
               "Average Purchase Price": avg_spend,
               "Total Expenditure": total_spend})

top_analysis= pd.DataFrame(top_analysis).sort_values(by=["Total Expenditure"],ascending=False)

top_analysis=top_analysis.style.format({"Average Purchase Price": "${:.2f}",
                                       "Total Expenditure": "${:.2f}"})

top_analysis




```
